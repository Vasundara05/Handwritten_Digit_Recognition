```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
#Loading tbe dataset
mnist=tf.keras.datasets.mnist
```

```python
#Split data into trin and test
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

```python
#Normalizing the data
x_train=tf.keras.utils.normalize(x_train,axis=1)
x_test=tf.keras.utils.normalize(x_test,axis=1)
```

```python
#Model creation
model=tf.keras.models.Sequential()
```

```python
#Adding layers
model.add(tf.keras.layers.Flatten(input_shape=(28,28)))
model.add(tf.keras.layers.Dense(128,activation="relu"))
model.add(tf.keras.layers.Dense(10,activation="softmax"))
```

```python
#Compiling the model
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
#Fit the model
model.fit(x_train,y_train,epochs=50)
model.save('handwrittendigit.h5')
```

```python
#Model metrics
loss,accuracy=model.evaluate(x_test,y_test)
print('Loss:',loss)
print('Accuracy:',accuracy)
```

```python
#Trying a classification with ral time image
im_n=1
while os.path.isfile(f"dig/dig{im_n}.png"):
    try:
        img=cv2.imread(f"dig/dig{im_n}.png")[:,:,0]
        img=np.invert(np.array([img]))
        prediction=model.predict(img)
        print(f"The number might be {np.argmax(prediction)}")
        plt.imshow(img[0],cmap=plt.cm.binary)
        plt.show()
    except:
        print("Error")
    finally:
        im_n+=1
```
